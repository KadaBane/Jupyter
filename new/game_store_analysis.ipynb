# Project description

This study looked at the video game market from 1980 to 2016.

In this study, an in-depth analysis of the gaming industry and the gaming platform industry was carried out.

An analysis of sales in 3 main regions was carried out: Europe, North America and Japan, 
as well as an analysis of sales trends and the popularity of certain platforms and genres in individual regions.

The analysis results are presented in the HTML format document in the Python language, 
in the Power Point presentation and there is also a link to an interactive dashboard made using Tabloau.

______
The project consists of 3 parts:

    Notebook which is an in depth analysis of the data
    
    Presentation of key findings and recommendations for an add campaign using the Power Point presentation 
    (Link: https://drive.google.com/file/d/1aQJx-2kYPVDAEXtTInrqywlvOvx7Hyvr/view?usp=sharing)
    
    Interactive Dashboard for monitoring changes over time 
    (Link: https://public.tableau.com/profile/daniel.katz4372#!/vizhome/VideoGamesSalesAnalysis_16148842421380/VideogamesSalesReport)
______
You work for the online store Ice, which sells video games all over the world. User and expert reviews, genres, platforms (e.g. Xbox or PlayStation), and historical data on game sales are available from open sources. You need to identify patterns that determine whether a game succeeds or not. This will allow you to spot potential big winners and plan advertising campaigns.

In front of you is data going back to 2016. It’s December 2016 and you’re planning a campaign for 2017.

The dataset contains the abbreviation ESRB. The Entertainment Software Rating Board evaluates a game's content and assigns an age rating such as Teen or Mature.
